```python
#podemos qualquer valor de um dos valores do describe
df1.Fare.describe()['75%']
```

```python
#determinando o interquartil
Q1 = df1.Fare.describe()['25%']
Q3 = df1.Fare.describe()['75%']
interquartil = Q3 -Q1

#calculando o valor mÃ¡ximo
vlr_max = Q3 + 1.5* interquartil
```

```python
#outliers nÃ£o sÃ£o valores errados e sim distantes
vlr_max
```

```python
df1[df1.Fare > vlr_max]
```
